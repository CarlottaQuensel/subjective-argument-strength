The data used to train a storytelling classifier and annotate the argument data is provided as part of:

Neele Falk and Gabriella Lapesa. 2022. Reports of personal experiences and stories in argumentation: datasets and analysis. In *Proceedings of the 60th Annual Meeting
of the Association for Computational Linguistics (Volume 1: Long Papers)*, pages 5530-5553, Dublin, Ireland. Association for Computational Linguistics.

and obtained from https://github.com/Blubberli/storytestimony/blob/main/datasets/whole_datasets.zip

```python
import pandas as pd
```

```python
cmv_raw = pd.read_csv("cmv_all_features.csv", sep="\t")
cmv_raw.head()
```

```python
europolis_raw = pd.read_csv("europolis_all_features.csv", sep="\t")
europolis_raw.head()
```

```python
regroom_raw = pd.read_csv("regroom_all_features.csv", sep="\t")
regroom_raw.head()
```

```python
cols = set(cmv_raw.columns)
cols.remove("post_text")
cols.remove("label")
cmv_only = cmv_raw.drop(columns=cols)
cmv_only.rename(columns={"post_text": "text"}, inplace=True)
cmv_only.to_csv("storytelling_cmv_only.csv", sep="\t", index=False)
```

```python
cols = set(regroom_raw.columns)
cols.remove("post_text")
cols.remove("label")
regroom_only = regroom_raw.drop(columns=cols)
regroom_only.rename(columns={"post_text": "text"}, inplace=True)
```

```python
cols = set(europolis_raw.columns)
cols.remove("post_text")
cols.remove("label")
europolis_only = europolis_raw.drop(columns=cols)
europolis_only.rename(columns={"post_text": "text"}, inplace=True)
```

```python
mixed = pd.concat((cmv_only, regroom_only, europolis_only), axis=0)
mixed = mixed.sample(frac=1)
mixed.head()
mixed.to_csv("storytelling_mixed.csv", sep="\t", index=False)
```
